#### What's the mean salary by department?
Hint : There are 2 ways to do it - one way is with groupby and the other way is with pivot tables. Do it both ways. 


```python
# Method 1
mean_salary_by_department = employee.groupby(['Department']).agg({'Salary': 'mean'}).reset_index().rename(columns={'Salary': 'Mean_salary'})
mean_salary_by_department

```

```python
# Method 2
# superstore.pivot_table(index= ['Category','Region'], values = ['Sales','Profit'], aggfunc = ['sum','mean'])
mean_salary_by_department_pivot = employee.pivot_table(index= ['Department'], values = ['Salary'], aggfunc = ['mean']).reset_index()
mean_salary_by_department_pivot
```

#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title.

```python
# Method 2
# your answer here
# superstore.pivot_table(index= ['Category','Region'], aggfunc = ['max', pd.Series.nunique, 'mean']).head(5)
mean_salary_by_department_title = employee.pivot_table(index= ['Department', 'Title'], aggfunc = {'Salary':'mean'})
mean_salary_by_department_title.rename(columns={'Salary': 'Mean_salary'}, inplace=True)
mean_salary_by_department_title

```

#### We want to display the number of years of work experience and the salary for each Title. Add Years as a column in your pivot table.

```python
# pd.pivot_table(employee, values=['Years', 'Salary'], index=['Title'], columns='Years', aggfunc='sum', fill_value=None, margins=False, dropna=True, margins_name='All', observed=False, sort=True)

years_salary_for_title = pd.pivot_table(employee, values=['Years', 'Salary'], index='Title', aggfunc='mean')
years_salary_for_title.rename(columns={'Salary': 'Mean_salary', 'Years': 'Mean_number_of_years'}, inplace=True)
years_salary_for_title
```

There is one Case Department Sales, Title associate, where max is equal to the min and the median.

#### The stake holders want to know for each department the number of employees and how much money is spend on salaries. Could you provide that information? 

```python
This_one2 = employee.pivot_table(index= ['Department'], values = ['Name', 'Salary'], aggfunc = {'Salary':'sum', 'Name':'count'}).reset_index()
This_one2.rename(columns={'Name': 'number_of_employees', 'Salary': 'Total_amount_spend_on_salaries'}, inplace=True)
This_one2
```

#### For each Department and Title, we want to know the the total years of work experience and the mean salary.

```python
This_one3 = employee.pivot_table(index= ['Department', 'Title'], values = ['Years', 'Salary'], aggfunc = {'Salary':'mean', 'Years':'sum'}).reset_index()
This_one3.rename(columns={'Years': 'Total_years_of_work_experience', 'Salary': 'Mean_salary'}, inplace=True)
This_one3
```
